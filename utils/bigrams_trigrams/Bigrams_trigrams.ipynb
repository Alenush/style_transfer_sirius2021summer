```python
def delete_rep(text):
  text = re.sub(' +', ' ', text)
  text = re.sub('(эй )+', 'эй ', text)
  text = re.sub('(э )+', 'э ', text)
  text = re.sub('(ха )+', 'ха ', text)
  text = re.sub('(размытие )+', 'размытие ', text)
  text = re.sub('(ох )+', 'ох ', text)
  text = re.sub('(хо )+', 'хо ', text)
  text = re.sub('(ой )+', 'ой ', text)
  text = re.sub('(подожди )+', 'подожди ', text)
  text = re.sub('(подождите )+', 'подождите ', text)
  text = re.sub('(ладно )+', 'ладно ', text)
  text = re.sub('(ла )+', 'ла ', text)
  text = re.sub('(смотри )+', 'смотри ', text)
  text = re.sub('(смотрите )+', 'смотрите ', text)
  text = re.sub('(ву )+', 'ву ', text)
  text = re.sub('(бла )+', 'бла ', text)
  text = re.sub('(хур )+', 'хур ', text)
  text = re.sub('(хорошо )+', 'хорошо ', text)
  text = re.sub('(погоди )+', 'погоди ', text)
  text = re.sub('(погодите )+', 'погодите ', text)
  text = re.sub('(ага )+', 'ага ', text)
  text = re.sub('(давай )+', 'давай ', text)
  text = re.sub('(хорошая игра )+', 'хорошая игра ', text)
  text = re.sub('(вонючий кот )+', 'вонючий кот ', text)
  text = re.sub('(7 )+', '7 ', text)
  text = re.sub('(друг )+', 'друг ', text)
  return text
```

```python
import pymorphy2

main_characters = ['ДЖОУИ', 'МОНИКА', 'РЕЙЧЕЛ', 'РОСС', 'ФИБИ', 'ЧЕНДЛЕР']

for name in main_characters:
  url = f'https://raw.githubusercontent.com/Alenush/style_transfer_sirius2021summer/main/data/scripts/russian/{name}.txt'
  req = requests.get(url)
  if req.status_code == requests.codes.ok:
      req = BeautifulSoup(req.text, 'html.parser')
      text = str(req)
  else:
      print('Content was not found.')
  text = re.sub(r'[^\w\s]', ' ', text).lower()#.replace('1230123', "'")
  text = delete_rep(text)
  text_list = text.split(' ')

  morph = pymorphy2.MorphAnalyzer(lang='ru')
  text_list = [morph.parse(el)[0][2] for el in text_list if el not in stopwords.words('russian') and el != '' and el not in stop_ru]
  text_list = [el for el in text_list if el not in stopwords.words('russian') and el != '' and el not in stop_ru and 'привет' not in el and 'черт' not in el and 'черта' not in el]
  text_list = [el if el != 'ладный' else 'ладно' for el in text_list]

  print(f"{name}.\n{Counter(text_list).most_common(10)}")
```

```python
main_characters = ['ДЖОУИ', 'МОНИКА', 'РЕЙЧЕЛ', 'РОСС', 'ФИБИ', 'ЧЕНДЛЕР']
dict_freq = dict()

for name in main_characters:
  url = f'https://raw.githubusercontent.com/Alenush/style_transfer_sirius2021summer/main/data/scripts/russian/{name}.txt'
  req = requests.get(url)
  if req.status_code == requests.codes.ok:
      req = BeautifulSoup(req.text, 'html.parser')
      text = str(req)
  else:
      print('Content was not found.')
  text = text.replace(".", ' 1234').replace("!", ' 1234').replace("?", ' 1234')
  text = re.sub(r'[^\w\s]', ' ', text).lower()#.replace('1230123', "'")
  
  text = delete_rep(text)

  text = text.split(' 1234')
  
  text = [el.strip() for el in text if el != '']
  bigrams_tmp = [b for l in text for b in zip(l.split(" ")[:-1], l.split(" ")[1:])]
  morph = pymorphy2.MorphAnalyzer(lang='ru')
  bigrams = [(morph.parse(el[0])[0][2], morph.parse(el[1])[0][2]) for el in bigrams_tmp]
  bigrams = [el for el in bigrams if el[0] not in stopwords.words('russian') and el[1] not in stopwords.words('russian') and el[0] not in stop_ru and el[1] not in stop_ru and 'привет' not in el and 'черт' not in el and 'черта' not in el]
  dict_freq[name] = Counter(bigrams).most_common(20)
  print(f"{name}.\n{Counter(bigrams).most_common(7)}")
```

```python
fig, axs = plt.subplots(3,2, figsize=(20,15))
fig.suptitle('Vertically stacked subplots')
i = 0
for fr in dict_freq:
  solo_freq = dict()
  for el in dict_freq[fr]:
    phr = ' '.join(el[0])
    if phr == 'черта взять':
      phr = 'черт возьми'
    elif phr == 'черта побрать':
      phr = 'черт побери'
    elif phr == 'послушать росс':
      phr = 'послушай росс'
    elif phr == 'нижний бельё':
      phr = 'нижнее бельё'
    elif phr == 'заниматься секс':
      phr = 'заниматься сексом'
    elif phr == 'идти сюда':
      phr = 'иди сюда'
    elif phr == 'ладный знать':
      phr = 'ладно знаю'
    elif phr == 'привет чендлера':
      phr = 'привет чендлер'
    elif phr == 'ой извинить':
      phr = 'ой извините'
    elif phr == 'ой подождать':
      phr = 'ой подожди'
    elif phr == 'друг друг':
      phr = 'друг друга'
    elif phr == 'крайний мера':
      phr = 'крайняя мера'
    elif phr == 'слава бог':
      phr = 'слава богу'
    solo_freq[phr] = el[1]
    
    
  cloud = WordCloud(max_font_size=80,colormap="hsv").generate_from_frequencies(solo_freq)
  axs[i%3, i%2].imshow(cloud, interpolation='bilinear')
  axs[i%3, i%2].axis('off')
  axs[i%3, i%2].set_title(fr, color='w')
  i += 1
plt.show()
```
