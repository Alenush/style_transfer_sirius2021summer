```python
#!g1.1
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
# 1. Init Lemmatizer
lemmatizer = WordNetLemmatizer()
# 2. Lemmatize Single Word with the appropriate POS tag
word = 'feet'
print(lemmatizer.lemmatize(word, get_wordnet_pos(word)))
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def remove_by_indices(x, indices):
    new_str = ""
    curr_index = 0
    end = 0
    for start, end in indices:
        while curr_index < start:
            new_str += x[curr_index]
            curr_index += 1
        curr_index = end + 1
    new_str += x[end:]
    return new_str

def get_prop_indices(s):
    lst = []
    doc = nlp(s)
    for token in doc.to_json()['tokens']:
        if token["pos"] == 'PROPN':
            lst.append((token['start'], token['end']))

    return lst
```

```python
#!g1.1
import re

def lemmatize(text):
    return [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(text)]

def base_form(text: str) -> str:
    char_only = re.sub('[^a-zA-Z ]', '', text) 
    new_str = str(re.sub(' +', ' ', char_only))
    lemmatized = ' '.join(lemmatize(new_str))
    without_names = remove_by_indices(lemmatized, get_prop_indices(lemmatized))
    return without_names.lower().strip()

text = "Yeah, but remember you said you ordered"
base_form(text)
```

```python
#!g1.1
suspisious_phrases = set(df_["Base_Phrase"].value_counts()[df_["Base_Phrase"].value_counts() > 1].index)
len(suspisious_phrases)
```

```python
#!g1.1
phrase_to_character = dict()
to_del = set()

for index, row in df_.iterrows():
    phrase = row["Base_Phrase"]
    hero = row["Character"]
    if phrase in suspisious_phrases:
        if phrase in phrase_to_character:
            if hero not in phrase_to_character[phrase]:
                phrase_to_character[phrase].append(hero)
        else:
            phrase_to_character[phrase] = [hero]

phrase_to_character
```

```python
#!g1.1
df_vc = df_["Base_Phrase"].value_counts()
df_vc = df_vc[df_vc > 1]
vc = df_vc.to_dict()
vc
```

```python
#!g1.1
df_vc = df_vc.reset_index()
df_vc.columns = ['Base_Phrase', 'Count']
df_vc
```

```python
#!g1.1
df_vc["Character"] = df_vc["Base_Phrase"].apply(lambda x: phrase_to_character[x])
df_vc
```

```python
#!g1.1
df_vc["Char_count"] = df_vc["Character"].apply(lambda x: len(x))
df_vc
```

```python
#!g1.1
unconfident_phrases = df_vc[(df_vc.Count >= 4) & (df_vc.Char_count > 2)].reset_index(drop=True)
unconfident_phrases
```

```python
unconfident_phrases["Word_count"] = unconfident_phrases["Base_Phrase"].progress_apply(lambda x: len(x.split()))
unconfident_phrases
```

```python
#!g1.1
bad_phr = list(unconfident_phrases["Base_Phrase"])
with open("unconfident_base_phrases.txt", "w") as f:
    f.write('\n'.join(bad_phr))

bad_phr
```
